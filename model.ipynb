# Load Yolov5 Model

```python
# load the pre trained ultralytics yolov5 model (baseline model)
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
```

# Make predictions with images

```python
img = 'https://ultralytics.com/images/zidane.jpg'
```

```python
results = model(img)
results.print()
```

```python
# render some images
%matplotlib inline

plt.imshow(np.squeeze(results.render()))
plt.show()
```

# Real time Detections

```python
# access the webcam and make detections
cap = cv2.VideoCapture(0) # if want to make predictions using stock videos, replace 0 with the local path to the video

while cap.isOpened():
    ret, frame = cap.read()
    
    # Make detections 
    results = model(frame)
    
    # cv2.imshow('YOLO', frame)
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

# Train the model from scatch with custom data

```python
import uuid   # Unique identifier
import os
import time
```

```python
# Create a directory for the training images
IMAGES_PATH = os.path.join('data', 'images') #/data/images
labels = ['awake', 'drowsy']
number_imgs = 20
```

```python
cap = cv2.VideoCapture(0)

# loop through labels
for label in labels:
    print('Collecting images for {}'.format(label))
    time.sleep(5)
    
    # loop through to capture images
    for image_num in range(number_imgs):
        print('Collecting images for {}, image number {}'.format(label, image_num))
        
        # Webcam feed
        ret, frame = cap.read()
        
        # See cell below - naming image path
        imgname = os.path.join(IMAGES_PATH, label+'.'+str(uuid.uuid1())+'.jpg')
        
        # write images to files
        cv2.imwrite(imgname, frame)
        
        # Render to the screen
        cv2.imshow('Image Collection', frame)
        
        # 2s delay between captures
        time.sleep(2)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
    
cap.release()
cv2.destroyAllWindows()
```

```python
# how the file is saved to the path
print(os.path.join(IMAGES_PATH, labels[0]+'.'+str(uuid.uuid1())+'.jpg'))
```
